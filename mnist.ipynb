```python
import networkx as nx
import pandas as pd

from writeNodesEdges import writeObjects
```

```python
with open('mnist_knn_10000.csv') as edges:
    next(edges)   # skip header
    graph: nx.Graph = nx.parse_edgelist(edges, delimiter=',', nodetype=int)
```

```python
df = pd.read_csv('mnist_10000.csv')
df
```

```python
def generate_points():
    # df[df.columns[:-1]] == skip last column of df
    for coords in df[df.columns[:-1]].values:
        yield list(coords)
```

```python
labels = list(df.label)
degrees = [graph.degree[node] for node in range(len(graph.nodes))]
```

```python
writeObjects(generate_points(), edges=graph.edges, scalar=labels, name='label', fileout='mnist_knn', scalar2=degrees,
             name2='degree')
```
